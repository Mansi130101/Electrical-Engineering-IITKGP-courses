````python
def get_ngrams_freqDist(n, ngramList):
    #This function computes the frequency corresponding to each ngram in ngramList 
    #Here, n=1 for unigram, n=2 for bigram, etc.
    #ngramList = list of unigrams when n=1, ngramList = list of bigrams when n=2
    #Returns: ngram_freq_dict (a Python dictionary where key = a ngram, value = its frequency)
    
    # *** Write code ***
    ngram_freq_dict = {}

    for num in ngramList:
        if num not in ngram_freq_dict.keys():
            ngram_freq_dict[num] = 1
        else:
            ngram_freq_dict[num] += 1
    
    return ngram_freq_dict
````

## **Next three words' Prediction using Smoothed Models**

For a bigram model, add-one smoothing is defined by $P_{Add-1}(w_i|w_{i-1})=\frac{count(w_{i-1},w_i)+1}{count(w_{i-1})+V}$.
That is, pretend we saw each word one more time than we did.

You have two tasks here.

First, compute the smoothed bigram and trigram models from the bigrams_freqDist and trigrams_freqDist you calculated above (use the unprocessed models). Second, using these smoothed models, predict the next 3 possible word sequences for testSent1, testSent2 and testSent3, using your smoothed models.

As for example, for the string 'Raj has a' the answers can be as below: 

(1) Raj has a **beautiful red car**

(2) Raj has a **charismatic magnetic personality**

````python
testSent1 = "There was a sudden jerk, a terrific convulsion of the limbs; and there he"
testSent2 = "They made room for the stranger, but he sat down"
testSent3 = "The hungry and destitute situation of the infant orphan was duly reported by"
````

````python
def smooth_bigram(unigrams_freqDist,bigrams_freqDist,sentence):
    prev_word = sentence.split(" ")[-1] 
    next_words = []; prob_list = []
    
    for _ in range(3):
        prob = 0; word_i = ""                                        
        for word, count_current in unigrams_freqDist.items():   
            prob_current = 0; count_prev_current = 0 
            
            if (prev_word, word) in bigrams_freqDist.keys():
                count_prev_current = bigrams_freqDist[(prev_word,word)]
            
            prob_current = (count_prev_current+1)/(count_current + len(unigrams_freqDist.keys()) )
            if  (prob_current>prob):
                prob = prob_current
                word_i = word

        next_words.append(word_i)
        prob_list.append(prob)
        prev_word = word_i
        
    return next_words 
````

````python
def smooth_trigram(unigrams_freqDist,bigrams_freqDist,trigrams_freqDist,sentence):
    prev_word1 = sentence.split(" ")[-1] 
    prev_word2 = sentence.split(" ")[-2] 
    next_words = []; prob_list = []
    
    for _ in range(3):
        prob = 0; word_i = ""                                        
        for word, count_current in unigrams_freqDist.items():   
            prob_current = 0; count_prev_prev = 0 
            
            if (prev_word2,prev_word1,word) in trigrams_freqDist.keys():
                count_prev_prev = trigrams_freqDist[(prev_word2,prev_word1,word)]
            else:
                count_prev_current=0
                
            if (prev_word2, prev_word1) in bigrams_freqDist.keys():
                count_prev_current = bigrams_freqDist[(prev_word2,prev_word1)]
            
            prob_current = (count_prev_prev+1)/(count_prev_current + len(unigrams_freqDist.keys()) )
            if  (prob_current>prob):
                prob = prob_current
                word_i = word

        next_words.append(word_i)
        prob_list.append(prob)
        prev_word2 = prev_word1
        prev_word1 = word_i
        
    return next_words
````

````python
def pred_bigram(sentence):
    bigram_next_n = smooth_bigram(unigrams_freqDist,bigrams_freqDist,sentence)
    bigram_pred_sent = sentence+" "+" ".join(bigram_next_n)
    return bigram_pred_sent

def pred_trigram(sentence):
    trigram_next_n = smooth_trigram(unigrams_freqDist,bigrams_freqDist,trigrams_freqDist,sentence)
    trigram_pred_sent = sentence+" "+" ".join(trigram_next_n)
    return trigram_pred_sent

#test sent 1
print("testSent1:",testSent1)
print("bigram -->",pred_bigram(testSent1))
print("trigram-->",pred_trigram(testSent1))

#test sent 2
print("\ntestSent2:",testSent2)
print("bigram -->",pred_bigram(testSent2))
print("trigram-->",pred_trigram(testSent2))

#test sent 3
print("\ntestSent3:",testSent3)
print("bigram -->",pred_bigram(testSent3))
print("trigram-->",pred_trigram(testSent3))
````

````python
def perplexity(model, sentence):
    prob = 1; perplexity = None
    words = sentence.split(" ")
    if model == "unigram":
        N=len(words)
        for w in words:
            count = 0
            if w in unigrams_freqDist.keys():
                count = unigrams_freqDist[w]
            prob = prob * (count+1)/(len(unigrams)+len(unigrams_freqDist.keys()))
    
    elif model == "bigram":
        N=0
        for i in ngrams(word_tokenize(sentence),2):
          N+=1
        for i in range(1,len(words)):
            word = words[i]
            prev_word = words[i-1]
            count_1 = 0
            if prev_word in unigrams_freqDist.keys():
                count_1 = unigrams_freqDist[prev_word] 
            count_2 = 0
            if (prev_word,word) in bigrams_freqDist.keys():
                count_2 = bigrams_freqDist[(prev_word,word)]
            prob = prob * (count_2+1)/(count_1+len(unigrams_freqDist.keys()))
        
    elif model == "trigram":
        N=0
        for i in ngrams(word_tokenize(sentence),3):
          N+=1
        for i in range(2,len(words)):
            word = words[i]
            prev_word = words[i-1]
            prev_prev_word = words[i-2]
            count_2 = 0
            if (prev_word,word) in bigrams_freqDist.keys():
                count_2 = bigrams_freqDist[(prev_word,word)]
            count_3 = 0
            if (prev_prev_word,prev_word,word) in trigrams_freqDist.keys():
                count_3 = trigrams_freqDist[(prev_prev_word,prev_word,word)]
            prob = prob * (count_3+1)/(count_2 + len(unigrams_freqDist.keys()))
    
    perplexity = (1/prob)**(1/N) 
    return perplexity
````
